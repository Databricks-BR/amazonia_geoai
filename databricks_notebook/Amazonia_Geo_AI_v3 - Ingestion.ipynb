```python
import geopandas as gpd
import urllib
import os

def download_file(url):
  filename = url.split('/')[-1] 
  response = urllib.request.urlopen(url)
  content = response.read()
  with open(filename, 'wb' ) as f:
      f.write( content )  
  return filename
  
def convert_shape_to_delta(url_shape_file, delta_table):

    geoshape = download_file(url_shape_file)
    assert(os.path.exists(geoshape))

    gdf_shape = gpd.read_file(geoshape)

    spark_df_geometry = spark.createDataFrame(gpd.GeoDataFrame(gdf_shape).to_wkt())

    spark_df_geometry.write.mode("overwrite").saveAsTable(delta_table)

```

```python

URL_SHAPE =  'https://github.com/Databricks-BR/amazonia_geoai/raw/main/maps/GEOFT_TERRA_INDIGENA.zip'
UC_DELTA_TABLE = f"amazonia.geo.bronze_shp_terra_indigena"

convert_shape_to_delta(URL_SHAPE, UC_DELTA_TABLE)
```

```python
%sql
SET spark.sql.shuffle.partitions = 200;

CREATE TABLE IF NOT EXISTS amazonia.geo.silver_h3_terra_indigena
AS
(
    SELECT explode(h3_polyfillash3(geometry, 10)) as h3_10_id
    FROM amazonia.geo.bronze_shp_terra_indigena
    WHERE terrai_cod = 31101
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS amazonia.geo.silver_h3_terra_indigena_v2
AS
(
  SELECT *, explode(h3_polyfillash3(geometry, 10)) as h3_10_id 
  FROM amazonia.geo.bronze_shp_terra_indigena
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS amazonia.geo.silver_h3_conservation_units
AS
(
  SELECT *, explode(h3_polyfillash3(geometry, 10)) as h3_10_id 
  FROM amazonia.geo.bronze_shp_conservation_units
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS amazonia.geo.silver_h3_queimadas
AS
(
  SELECT *, h3_longlatash3(lat,lon,10) as h3_10_id 
  FROM amazonia.geo.bronze_shp_queimadas
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS amazonia.geo.silver_h3_desmatamento
AS
(
  SELECT *, explode(h3_polyfillash3(geometry, 10)) as h3_10_id 
  FROM amazonia.geo.bronze_shp_desmatamento
);
```

```python
%sql
CREATE FUNCTION IF NOT EXISTS amazonia.geo.impact_score(area_conservacao INT, numero_dias_sem_chuva INT, terra_indigena INT) RETURNS INT
  RETURN area_conservacao + numero_dias_sem_chuva + terra_indigena;
```

```python
%sql
CREATE FUNCTION IF NOT EXISTS amazonia.geo.critical_score(area INT, area_desmatamento INT, risco_fogo INT) RETURNS INT
  RETURN area + area_desmatamento + risco_fogo;
```
